## Exploring News API

```python
import requests
```

```python
# visit newsapi.org to get API key

url = ('https://newsapi.org/v2/everything?'
       'q=DeepSeek&'
    #    'from=2025-02-07&'
       'sortBy=popularity&'
       'apiKey=YOUR_NEWS_API_KEY')

response = requests.get(url)
response
```
